```python
import pandas as pd

df = pd.read_csv('collect1.csv')
df.head(3)
```

```python
import preprocessor as p

def preprocess_tweet(row):
    text = row['text']
    text = p.clean(text)
    return text

df['text'] = df.apply(preprocess_tweet, axis = 1)
```

```python
text = df['text']
tweets1 = []

for i in text:
    tweets1.append(i)
print(tweets1)
```

```python
tweets2 = []
for i in tweets1:
    input_str = i
    input_str = input_str.lower()
    tweets2.append(input_str)
```
